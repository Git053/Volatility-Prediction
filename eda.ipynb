```python
import pandas as pd
import numpy as np
import plotly.express as px
train = pd.read_csv('/content/drive/MyDrive/data/orderbook/train.csv')
train = train.drop(labels=range(1001,592380), axis=0)
train
```

```python
train['wap'] = (train['bid1'] * train['ask1vol'] +
                                train['ask1'] * train['bid1vol']) / (
                                       train['bid1vol']+ train['ask1vol'])
```

```python
fig = px.line(train, x="id", y="wap", title='WAP of stock_id_0, time_id_5')
fig.show()
```

```python
import matplotlib.pyplot as plt
```

```python
samples = ["bid1","bid2","ask1","ask2"]

for num,a in enumerate(samples):
    plt.figure(figsize=(20,5))
   
    plt.subplot(4,1,num+1)
    plt.plot(train["id"],train[a])
    plt.title(a)
plt.show()
plt.figure(figsize=(20,5))

for num,a in enumerate(samples):
    
   
    plt.plot(train["id"],train[a],label=a)
plt.legend(fontsize=12)
```
